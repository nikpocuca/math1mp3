```python
# differential equations 

#lotka voltera equations 

# x are the rabbits 
# y are the wolves 

# change in rabbit population wrt time. 
# prey pop
# dx/dt = a*x−b*x*y    

# change in wolf population wrt time. 
# predator pop
# dy/dt = r*x*y − c*y

h = 0.001

wolf_pop = np.array([2])
rabbit_pop = np.array([3])

a = 1 # increase in rabbits naturally
b = 1 # influence of wolves eating rabbits
r = 1 #  wolves increase population para
c = 1 # wolves decay

# xnew = xold + time_step * change of x 

for i in range(0,20000):
    rabbit_change = a * rabbit_pop[i] - b * rabbit_pop[i] * wolf_pop[i]
    wolf_change = r * rabbit_pop[i] * wolf_pop[i] - c * wolf_pop[i]

    wolf_pop_now = wolf_pop[i] + h*wolf_change
    rabbit_pop_now = rabbit_pop[i] + h*rabbit_change
    wolf_pop = np.append(wolf_pop,wolf_pop_now)
    rabbit_pop = np.append(rabbit_pop,rabbit_pop_now)
    # print("wolf change",wolf_change)
    # print("rabbit change",rabbit_change)
    # print("wolf pop", wolf_pop[i+1], "rabbit pop",rabbit_pop[i+1])

t = np.linspace(0,20001,num = 20001)

pyplot.plot(t,wolf_pop, label = "wolves")
pyplot.plot(t,rabbit_pop, label = "rabbits")
pyplot.legend()


```

```python


# radioactive material, how much of is there, (weight in kg)... and what is its half-life? 
# half life means, how long does 1kg take to decay to 0.5kg? 
# 250kg , 125kg... = halflife
# my halflife is 5 years. 
# 250kg take to reach 10kg. 

plutonium = np.array([567])

# 1kg of plutonium 
# 24100 
# 1/24100

# dp/dt = -k*p_t

h = 1

# 1kg -> 0.5 kg of plutonium
# 24100 years. 

k = -(np.log(0.5/1))/24100 


#k = 0.07

for i in range(0,100000): 
    change_plutonium = -k*plutonium[i]
    plutonium_now = plutonium[i] + h*change_plutonium
    plutonium = np.append(plutonium,plutonium_now)

x = np.linspace(0,len(plutonium),num = len(plutonium))
pyplot.plot(x,plutonium)

```
