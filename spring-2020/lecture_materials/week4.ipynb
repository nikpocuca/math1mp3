```python
# differential equations 

#lotka voltera equations 

# x are the rabbits 
# y are the wolves 

# change in rabbit population wrt time. 
# prey pop
# dx/dt = a*x−b*x*y    

# change in wolf population wrt time. 
# predator pop
# dy/dt = r*x*y − c*y

h = 0.001

wolf_pop = np.array([2])
rabbit_pop = np.array([3])

a = 1 # increase in rabbits naturally
b = 1 # influence of wolves eating rabbits
r = 1 #  wolves increase population para
c = 1 # wolves decay

# xnew = xold + time_step * change of x 

for i in range(0,20000):
    rabbit_change = a * rabbit_pop[i] - b * rabbit_pop[i] * wolf_pop[i]
    wolf_change = r * rabbit_pop[i] * wolf_pop[i] - c * wolf_pop[i]

    wolf_pop_now = wolf_pop[i] + h*wolf_change
    rabbit_pop_now = rabbit_pop[i] + h*rabbit_change
    wolf_pop = np.append(wolf_pop,wolf_pop_now)
    rabbit_pop = np.append(rabbit_pop,rabbit_pop_now)
    # print("wolf change",wolf_change)
    # print("rabbit change",rabbit_change)
    # print("wolf pop", wolf_pop[i+1], "rabbit pop",rabbit_pop[i+1])

t = np.linspace(0,20001,num = 20001)

pyplot.plot(t,wolf_pop, label = "wolves")
pyplot.plot(t,rabbit_pop, label = "rabbits")
pyplot.legend()


```

```python
# okay so we can generate uniform numbers between (0,1)... so what? 

# well actually you can do a lot with this. 

# lets use a dice example. 

# i can show you how you can get dice throws (6 sided dice)
# from numbers between (0,1). 
```

```python
# what we can do is we can divide up the (0,1) interval in 6 even areas like so. 

#     Zone 1     Zone 2     Zone 3     Zone 4      Zone 5        Zone 6
#   [0.......1/6.......2/6........3/6.........4/6...........5/6............1]     
```

```python
1/6 # this is the first area, random numbers that fall between (0, 1/6) should be a 1 
2/6 # this is the second area, random numbers that fall between (1/6,2/6) should be a 2
3/6 # this is the third area, random numbers that fall between (2/6,1/2) should be a 3 
# and so on.... 
```

```python
# we need to include bounds, so for instance I will have geq for all cases like so: 

# [0,1/6)
# [1/6,2/6)
# [2/6,1/2) 
# [1/2, 4/6)
# [4/6,5/6)
# [5/6, 1]

# that way there are no overlaps between the zones. 

```

```python
# lets code this up! 

def d6(size):
    """we are going to pass in a size variable"""
    # lets generate our random uniform numbers. 
    x = random.uniform(0,1,size=size)

    result = np.array([])

    # now we are going to iterate through x and append to an np array result 
    for entry in x:
        if 0 <= entry < 1/6:
            result = np.append(result,1)
        elif 1/6 <= entry < 2/6:
            result = np.append(result,2)
        elif 2/6 <= entry < 1/2:
            result = np.append(result,3)
        elif 3/6 <= entry < 4/6:
            result = np.append(result,4)
        elif 4/6 <= entry < 5/6:
            result = np.append(result,5)
        else:
            result = np.append(result,6)

    return(result)


```

```python
# Example 1 A 1500 gallon tank initially contains 600 gallons of water with 5 lbs of salt dissolved in it. Water enters the tank at a rate of 9 gal/hr and the water entering the tank has a salt concentration of 1/5 * ( 1 + cos(t)) lbs/gal. If a well mixed solution leaves the tank at a rate of 6 gal/hr, how much salt is in the tank when it overflows?


import numpy as np 
```

```python
# Initial Value problem. 
iw = 600
isalt = 5

t = np.linspace(0,20,num = 100)

# 9 gallons an hour.
#1/5*( 1 + np.cos(t)) 

iQ = isalt/iw

def dQ_dt(t):
    solution_in = 9*1/5*( 1 + np.cos(t)) 
    solution_out = 6*1/5*( 1 + np.cos(t))
    return (solution_in - solution_out)

h = 0.2020202

Q = np.array([iQ])
Water = np.array([iw])

for i in range(0,len(t)):
    # calculate change in concentration 
    Q_now = Q[i]
    Q_new = Q_now + h*dQ_dt(t[i])
    Q = np.append(Q,Q_new)
    # calculate change in water
    W_now = Water[i]
    water_change = 3*t[i]
    W_new = W_now + water_change
    Water = np.append(Water,W_new)
    # when it overflows
    if np.isclose(W_new,1500):
        break 


print(Q)
```

```python
def dIP_dt(IP):
    #natural increase 
    nat_inc = k*IP
    out_fact = -8 
    return(nat_inc + out_fact)
```

```python


def kinematics(deg,force,x_start,y_start):

    deg_rad = np.radians(deg)

    v_x = force*np.cos(deg_rad)
    v_y = force*np.sin(deg_rad)

    def dx_dt():
        return(v_x)

    def dy_dt(t_i):
        return(v_y + gravity*t_i)

    x_pos = np.array([x_start])
    y_pos = np.array([y_start])

    t = np.linspace(0,10,num = 100)
    h = t[1] - [0]

    for i in range(0,len(t)):
        x_now = x_pos[i]
        y_now = y_pos[i]

        change_x = dx_dt()
        change_y = dy_dt(t[i])
        
        x_new = x_now + h*change_x
        y_new = y_now + h*change_y

        x_pos = np.append(x_pos,x_new)
        y_pos = np.append(y_pos,y_new)
        
        if (y_new < 0):
            break 

        
    pyplot.plot(x_pos,y_pos)


```

```python


def kinematics_wind(deg,force,x_start,y_start,wind):

    deg_rad = np.radians(deg)

    v_x = force*np.cos(deg_rad)
    v_y = force*np.sin(deg_rad)

    def dx_dt(t_i):
        return(v_x  + wind*t_i )

    def dy_dt(t_i):
        return(v_y + gravity*t_i)

    x_pos = np.array([x_start])
    y_pos = np.array([y_start])

    t = np.linspace(0,10,num = 100)
    h = t[1] - [0]

    for i in range(0,len(t)):
        x_now = x_pos[i]
        y_now = y_pos[i]

        change_x = dx_dt(t[i])
        change_y = dy_dt(t[i])
        
        x_new = x_now + h*change_x
        y_new = y_now + h*change_y

        x_pos = np.append(x_pos,x_new)
        y_pos = np.append(y_pos,y_new)
        
        if (y_new < 0):
            break 

        
    pyplot.plot(x_pos,y_pos)


```
