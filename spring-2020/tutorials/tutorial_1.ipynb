```python
# biological sex variable 
bio_sex = "F"
# age variable 
age = 42
# chol variable 
chol_var = 275
# smoker variable
is_smoker = True
# HDL CHOLESTOROL VARIABLE 
hdl_chol_var = 37
# blood pressure . 
sys_bp_var = 145
# treated or untreated 
is_treated = True


points = 0.0

if bio_sex == "F":
    if 20 <= age <= 34:
        points -= 7
    elif 35 <= age <= 39:
        points -= 3
    elif 45 <= age <= 49:
        points += 3
    elif 50 <= age <= 54:
        points += 6
    elif 55 <= age <= 59:
        points += 8
    elif 60 <= age <= 64:
        points += 10
    elif 65 <= age <= 69:
        points += 12
    elif 70 <= age <= 74:
        points += 14 
    elif 75 <= age <= 79:
        points += 16



    # Total cholesterol, mg/dL: Age 20–39 years: Under 160: 0 points. 160-199: 4 points. 200-239: 8 points. 
    # 240-279: 11 points. 280 or higher: 13 points. 
    if 20 <= age <= 39:
        if 160 <= chol_var <= 199:
            points += 4
        elif 200 <= chol_var <= 239:
            points += 8
        elif 240 <= chol_var <= 279:
            points += 11
        elif 280 <= chol_var:
            points += 13


    # Age 40–49 years: Under 160: 0 points. 160-199: 3 points. 200-239: 6 points. 240-279: 8 points.
    # 280 or higher: 10 points. • 
    if 40 <= age <= 49:
        if 160 <= chol_var <= 199:
            points += 3
        elif 200 <= chol_var <= 239:
            points += 6
        elif 240 <= chol_var <= 279:
            points += 8
        elif 280 <= chol_var:
            points += 10
    

    # Age 50–59 years: Under 160: 0 points. 160-199: 2 points. 200-239: 4 points. 240-279: 
    # 5 points. 280 or higher: 7 points. 
    if 50 <= age <= 59:
        if 160 <= chol_var <= 199:
            points += 2
        elif 200 <= chol_var <= 239:
            points += 4
        elif 240 <= chol_var <= 279:
            points += 5
        elif 280 <= chol_var:
            points += 7 

        
    # Age 60–69 years: Under 160: 0 points. 160-199: 1 point. 200-239:
    # 2 points. 240-279: 3 points. 280 or higher: 4 points. 
    if 60 <= age <= 69:
        if 160 <= chol_var <= 199:
            points += 1
        elif 200 <= chol_var <= 239:
            points += 2
        elif 240 <= chol_var <= 279:
            points += 3
        elif 280 <= chol_var:
            points += 4


    # Age 70–79 years: Under 160: 0 points. 160-199: 1 point. 200-239:
    # 1 point. 240-279: 2 points. 280 or higher: 2 points.
    if 70 <= age <= 79:
        if 160 <= chol_var <= 199:
            points += 1
        elif 200 <= chol_var <= 239:
            points += 1
        elif 240 <= chol_var <= 279:
            points += 2
        elif 280 <= chol_var:
            points += 2     


    # If cigarette smoker: Age 20–39 years: 9 points. • Age 40–49 years: 
    # 7 points. • Age 50–59 years: 4 points. • Age 60–69 years: 2 points. • Age 70–79 years: 1 point.  

    if is_smoker:
        if 20 <= age <= 39:
            points += 9
        elif 40 <= age <= 49:
            points += 7
        elif 50 <= age <= 59:
            points += 4
        elif 60 <= age <= 69:
            points += 2
        elif 70 <= age <= 79:
            points += 1


    # HDL cholesterol, mg/dL: 60 or higher: Minus 1 point. 50-59: 0 points. 40-49: 1 point. Under 40: 2 points.
    if 60 <= hdl_chol_var:
        points -= 1
    elif 40 <= hdl_chol_var <= 49:
        points += 1
    elif hdl_chol_var < 40:
        points += 2

    # Systolic blood pressure, mm Hg: Untreated: Under 120: 0 points. 120-129: 1 point. 130-139:
    # 2 points. 140-159: 3 points. 160 or higher: 4 points. • 
    if not is_treated:
        if 120 <= sys_bp_var <= 129:
            points += 1
        elif 130 <= sys_bp_var <= 139:
            points += 2
        elif 140 <= sys_bp_var <= 159:
            points += 3
        elif 160 <= sys_bp_var:
            points += 4
    if is_treated:
        if 120 <= sys_bp_var <= 129:
            points += 3
        elif 130 <= sys_bp_var <= 139:
            points += 4
        elif 140 <= sys_bp_var <= 159:
            points += 5
        elif 160 <= sys_bp_var:
            points += 6


    #Treated: Under 120: 0 points. 120-129: 3 points.
    # 130-139: 4 points. 140-159: 5 points. 160 or higher: 6 points.   


    # 10-year risk in %: Points total: Under 9 points: <1%. 9-12 points: 1%. 13-14 points: 2%. 15 points: 
    # 3%. 16 points: 4%. 17 points: 5%. 18 points: 6%.
    # 19 points: 8%. 20 points: 11%. 21=14%, 22=17%, 23=22%, 24=27%, >25= Over 30%


    if points < 9:
        print("You have less than 1 percent risk ")
    elif 9 <= points <= 12:
        print("You have 1 percent risk of cardio vascular disease")
    elif 13 <= points <= 14:
        print("You have 2 percent risk of cardio vascular disease")
    elif points == 15:
        print("You have 3 percent risk of cardio vascular disease")
    elif points == 16:
        print("You have 4 percent risk of cardio vascular disease")
    elif points == 17:
        print("You have 5 percent risk of cardio vascular disease")
    elif points == 18:
        print("You have 6 percent risk of cardio vascular disease")
    elif points == 19:
        print("You have 8 percent risk of cardio vascular disease")
    elif points == 20:
        print("You have 11 percent risk of cardio vascular disease")
    elif points == 21:
        print("You have 14 percent risk of cardio vascular disease")
    elif points == 22:
        print("You have 17 percent risk of cardio vascular disease")
    elif points == 23:
        print("You have 22 percent risk of cardio vascular disease")
    elif points == 24:
        print("You have 27 percent risk of cardio vascular disease")
    elif 25 <= points:
        print("You have over 30 percent risk of cardio vascular disease")


print("POINTS: ", points)

```
