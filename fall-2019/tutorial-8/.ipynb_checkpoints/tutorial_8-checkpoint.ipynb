```python
def stock_random_walk(s0,mu,p,n,random_generator):
        stock_pos = np.array([s0])
        for i in range(0,n):
            # simulated random uniform
            p_i = random_generator()
            sI = stock_pos[-1] + mu
            if sI <= 0:
                break
            stock_pos = np.append(stock_pos,sI)
        return (stock_pos)
```

```python
path_1 = stock_random_walk(50,1,1,0.5,50)
```

```python
for i in range(0,5):
    path_1 = stock_random_walk(50,1,1,0.5,1000)
    pyplot.plot(path_1)
```
