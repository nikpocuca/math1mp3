```python
def stock_random_walk(s0,mu,n,random_generator):
        stock_pos = np.array([s0])
        for i in range(0,n):
            # simulated random uniform
            p_i = random_generator(0,3)
            sI = stock_pos[-1] + mu*p_i
            if sI <= 0:
                break
            stock_pos = np.append(stock_pos,sI)
        return (stock_pos)
```

```python
path_1 = stock_random_walk(50,1,100,np.random.normal)
pyplot.plot(path_1)
```

```python
for i in range(0,5):
    path_1 = stock_random_walk(50,2,1000,np.random.normal)
    pyplot.plot(path_1)
```

```python
def TwoD_Random(n):
    xpos = np.array([0])
    ypos = np.array([0])
    for i in range(0,n):
        p = np.random.uniform(0,1)
        if p < 0.25:
            xpos = np.append(xpos,xpos[-1] + np.random.lognormal(1,2))
            ypos = np.append(ypos,ypos[-1])
        elif 0.25 <= p < 0.5:
            xpos = np.append(xpos,xpos[-1])
            ypos = np.append(ypos,ypos[-1] + np.random.lognormal(1,2))
        elif 0.50 <= p < 0.75:
            xpos = np.append(xpos,xpos[-1] - np.random.lognormal(1,2))
            ypos = np.append(ypos,ypos[-1])
        else:
            xpos = np.append(xpos,xpos[-1])
            ypos = np.append(ypos,ypos[-1] - np.random.lognormal(1,2))
    return ((xpos,ypos))            
```
