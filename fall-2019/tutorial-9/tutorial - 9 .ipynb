```python
theta = np.linspace(0,10*np.pi,10000)
```

```python
theta
```

```python
L = np.sin(theta*np.pi)
```

```python
x = np.cos(theta)*L
y = np.sin(theta)*L
```

```python
pyplot.plot(x,y)
pyplot.plot(x,-y)
pyplot.title("Clover pi")
pyplot.xlabel("x")
pyplot.ylabel("y")
```

```python
pyplot.scatter(x,y)
```
