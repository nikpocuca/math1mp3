```python
def SimulateStocks():
    sum_stock = [0]
    sum_stock += [rollStock()]
    for roll in range(1,101):
        sum_stock += [sum_stock[roll]+ random.normalvariate(2, 4)]
        
    return sum_stock
```

```python
manyStocks = []

for i in range(1,1001):
    manyStocks += [SimulateStocks()]

```

```python
def LineOfBF(x,y):
    X = numpy.matrix([numpy.array(x),numpy.array(ones)])
    b = (y*X.transpose()).transpose()
    A = X*X.transpose()
    coef = numpy.linalg.solve(A,b)
    return coef
```
